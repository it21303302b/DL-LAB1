```python
import numpy as np

# Creating a 4x4 array from an exponential distribution
array_4x4 = np.random.exponential(scale=1.0, size=(4, 4))
print(array_4x4)
```

```python
import matplotlib.pyplot as plt

# Creating a random 100000x1 array from an exponential distribution
exp_data = np.random.exponential(scale=1.0, size=100000)
uniform_data = np.random.uniform(low=0.0, high=1.0, size=100000)
normal_data = np.random.normal(loc=0.0, scale=1.0, size=100000)

# Plotting histograms
plt.hist(exp_data, density=True, bins=100, histtype="step", color="blue", label="Exponential")
plt.hist(uniform_data, density=True, bins=100, histtype="step", color="green", label="Uniform")
plt.hist(normal_data, density=True, bins=100, histtype="step", color="red", label="Normal")
plt.axis([-2.5, 5, 0, 1.1])
plt.legend(loc="upper right")
plt.title("Random distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Creating the grid
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = x**2 + y**2

# Plotting the 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')

ax.set_title('Z = X^2 + Y^2')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
tips = sns.load_dataset("tips")

# Select the features (substitute for similar numeric features in the tips dataset)
features = tips[['total_bill', 'tip', 'size']]

# Calculate the Pearson and Spearman correlation matrices
pearson_corr = features.corr(method='pearson')
spearman_corr = features.corr(method='spearman')

# Plot heatmaps
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', cbar=False)
plt.title('Pearson Correlation')

plt.subplot(1, 2, 2)
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', cbar=False)
plt.title('Spearman Correlation')

plt.tight_layout()
plt.show()
```
